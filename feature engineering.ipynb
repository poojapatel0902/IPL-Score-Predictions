```python
import pandas as pd
import numpy as np
import pickle
```

```python
df = pickle.load(open('dataset_level2.pkl','rb'))
```

```python
df
# batting_team
# bowling team
# city
# current score
# ball left
# wickets left
# current rr
# last five
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['venue'], inplace=True)
```

```python
df
```

```python
df['city'].value_counts()
```

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts()>600].index.tolist()
```

```python
df=df[df['city'].isin(eligible_cities)]
```

```python
df
```

```python
# Convert runs to numbers 
df['runs'] = pd.to_numeric(df['runs'], errors='coerce').fillna(0).astype(int)
```

```python
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df
```

```python
df = df.copy()
df['overs']=df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no']=df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python
df['overs'] = df['overs'].astype(int)
df['ball_no'] = df['ball_no'].astype(int)
```

```python
df['balls_bowled'] = (df['overs'] * 6) + df['ball_no']
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x: 0 if x < 0 else x)
df
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
df['current_run_rate'] = (df['current_score']*6)/df['balls_bowled']
df
```

```python
df['last_5_over'] = df.groupby('match_id')['runs'].transform(lambda x: x.rolling(window=30, min_periods=1).sum())
```

```python
df
```

```python
final_df=df.groupby('match_id').sum()['runs'].reset_index().merge(df, on='match_id')
final_df
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','current_run_rate','last_5_over','runs_x']]
final_df
```

```python
final_df.isnull().sum()
```

```python
final_df=final_df.sample(final_df.shape[0])
final_df
```
